```python
import os

from crewai import Agent,Task,Crew,Process
from crewai_tools import SerperDevTool
from langchain_groq import ChatGroq

import os
os.environ["SERPER_API_KEY"] = "e7039d4890868ff62a192312049e83e56ff6048f"  # serper.dev API key

print(os.environ["SERPER_API_KEY"])
os.environ['GROQ_API_KEY'] = "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"
```

```python
llm = ChatGroq(model="gemma-7b-it",groq_api_key=os.environ['GROQ_API_KEY'])
print(llm.invoke("hi"))
search_tool = SerperDevTool()
```

```python
Chatbot = Agent(
    role = "Chatbot",
    goal='Answer customer queries quickly and easily.',
    verbose=True,
    backstory=(
    "You are a skilled chatbot with a talent of solving customers query."
    "You can cover a large portion of customer inquires."
    "You provide 24/7 support to our customers."
    ),
    tools = [search_tool],
    llm = llm,
    max_iter = 5,
    allow_delegation=False
)
```

```python
write_task = Task(
    description=(
    "handle a simple requests for FAQs or Password Reset or"
    "perform more complex business tasks, like renewing a"
    "customer's membership or handling a refund."
  ),
    expected_output='A blog article on recent advancements in {subject_area} in markdown.',
    tools = [search_tool],
    agent = Chatbot,
    async_execution = False,
    output_file = "blog.md"
)
```

```python
crew = Crew(
    agents = [Chatbot],
    tasks = [write_task],
    process = Process.sequential,
    max_rpm = 3,
    cache=True
)

result = crew.kickoff(inputs={'subject_area':"Genrative AI"})
print(result)
```
